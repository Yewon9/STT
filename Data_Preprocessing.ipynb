# 일반 음성 데이터

## aihub_4

```python
aihub_4 = pd.read_csv('/content/drive/MyDrive/project/data_aihub/4.csv', encoding = 'CP949')
aihub_4
```

```python
aihub_4 = aihub_4.iloc[:, 0:2]
aihub_4 = aihub_4.rename(columns = {'wav_id' : 'Path', '발화문' : 'Text'})
aihub_4['Path'] = '/content/drive/MyDrive/project/data_aihub/4/' + aihub_4['Path'] + '.wav'
aihub_4
```

## aihub_5

```python
aihub_5 = pd.read_csv('/content/drive/MyDrive/project/data_aihub/5.csv', encoding = 'CP949')
aihub_5
```

```python
aihub_5 = aihub_5.iloc[:, 0:2]
aihub_5 = aihub_5.rename(columns = {'wav_id' : 'Path', '발화문' : 'Text'})
aihub_5['Path'] = '/content/drive/MyDrive/project/data_aihub/5/' + aihub_5['Path'] + '.wav'
aihub_5
```

## aihub_5_2

```python
aihub_5_2 = pd.read_csv('/content/drive/MyDrive/project/data_aihub/5_2.csv', encoding = 'CP949')
aihub_5_2
```

```python
aihub_5_2 = aihub_5_2.iloc[:, 0:2]
aihub_5_2 = aihub_5_2.rename(columns = {'wav_id' : 'Path', '발화문' : 'Text'})
aihub_5_2['Path'] = '/content/drive/MyDrive/project/data_aihub/5_2/' + aihub_5_2['Path'] + '.wav'
aihub_5_2
```

## general_wav

```python
general = pd.concat([aihub_4, aihub_5, aihub_5_2], axis = 0)
general
```

```python
general = general.drop_duplicates(subset = 'Text')
general.shape
```

```python
general = general[general['Path'].apply(lambda x: os.path.exists(x))]
general.shape
```

```python
general.to_csv('general.csv', index = False, encoding = 'utf-8-sig')
```
