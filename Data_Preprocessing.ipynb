### 15

```python
wav_dir = '/content/drive/MyDrive/project/data/Training/wav/15'
json_dir = '/content/drive/MyDrive/project/data/Training/json/15'
```

```python
def create_dataframe(wav_dir, json_dir):
    data = []
    for wav_file in os.listdir(wav_dir):
        if wav_file.endswith(".wav"):
            wav_path = os.path.join(wav_dir, wav_file)
            json_path = os.path.join(json_dir, wav_file.replace(".wav", ".json"))

            if os.path.exists(json_path):
                with open(json_path, 'r', encoding = 'utf-8') as f:
                    json_data = json.load(f)
                    text = json_data.get('Transcript', '')

                data.append({"Path" : wav_path, "Text" : text})

    return pd.DataFrame(data)
```

```python
disorder_15 = create_dataframe(wav_dir, json_dir)
disorder_15
```
